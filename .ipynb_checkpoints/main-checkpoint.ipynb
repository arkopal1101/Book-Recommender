## Collaborative Filtering

```python
book_ratings = rating.merge(books, on='ISBN')
```

```python
book_ratings.head()
```

```python
users_df = book_ratings.groupby('User-ID')['Book-Rating'].count() > 200
```

```python
users_df = users_df[users_df].index
```

```python
filtered_ratings = book_ratings[book_ratings['User-ID'].isin(users_df)]
```

```python
books_df = filtered_ratings.groupby('Book-Title')['Book-Rating'].count() >= 50
```

```python
books_df = books_df[books_df].index
```

```python
filtered_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(books_df)]
```

```python
filtered_ratings
```

```python
filtered_ratings.drop_duplicates()
```

```python
final_ratings = filtered_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
final_ratings.fillna(0, inplace=True)
```

```python
final_ratings
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(final_ratings)
```

```python
similarity_score.shape
```

```python
def recommend(book):
    index = np.where(final_ratings.index == book)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])), key= lambda x:x[1], reverse=True)[1:6]
    book_indexes = [i[0] for i in similar_books]
    books = [final_ratings.index[i] for i in book_indexes]
    
    return books
```

```python
recommend('You Belong To Me')
```
